```python
import matplotlib.pyplot as plt
import numpy as np
from typing import List, Union, Optional
```

```python
norm = np.random.normal(size=5000)
```

```python
def plot(x:np.ndarray, edges:Optional[Union[np.ndarray,List[float]]]=None) -> None:
    fig, axs = plt.subplots(1,2, figsize=(12,4))
    axs[0].hist(x, bins=edges, density=False)
    axs[1].hist(x, bins=edges, density=True)
    plt.show()
```

```python
def get_edges(x:np.ndarray, max_unc:float) -> np.ndarray:
    min_n = int(np.ceil(max_unc**-2))
    x.sort()
    edges = x[np.arange(0, len(x), min_n)]
    edges[-1] = x[-1]
    return edges
```

```python
plot(norm)
```

```python
edges = get_edges(norm, 0.05); len(edges), edges[0], edges[-1]
```

```python
plot(norm, edges)
```

```python
exp = np.random.exponential(scale=3, size=5000)-5
```

```python
data = np.hstack((norm, exp))
```

```python
edges = get_edges(data, 0.05); len(edges), edges[0], edges[-1]
```

```python
plot(data, edges)
```
